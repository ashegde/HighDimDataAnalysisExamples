# ADMM for MRI

In this notebook, we implement ADMM for MRI. These contents are based on Chapter 10.4 of __{ Wright, John, and Yi Ma. High-dimensional data analysis with low-dimensional models: Principles, computation, and applications. Cambridge University Press, 2022.}

Here we use the ADMM algorithm described in Eq. (10.4.13) of the main text.

__while__ $\boldsymbol x_k$ not converged (k=1,2, ...) __do__ <br>
>$\boldsymbol x^{(k+1)} \leftarrow \arg \min_x g_1(\boldsymbol z^{(k)}, \boldsymbol x, \boldsymbol \lambda_1^{(k)})$<br>
$\boldsymbol v^{(k+1)} \leftarrow \arg \min_{v_i} g_2(\boldsymbol z^{(k)}, \boldsymbol v_i, \boldsymbol \lambda_2^{(k)})$<br>
$\boldsymbol z^{(k+1)} \leftarrow \arg \min_{z} L(\boldsymbol z, \boldsymbol x^{(k+1)}, \boldsymbol v^{(k+1)}, \boldsymbol \lambda_1^{(k)}, \lambda_2^{(k)})$<br>
$\boldsymbol \lambda_1^{(k+1)} \leftarrow \boldsymbol \lambda_1^{(k)} + \mu_1(\boldsymbol x^{(k+1)}- \Phi \boldsymbol z^{(k+1)})$ <br>
$\boldsymbol \lambda_2^{(k+1)} \leftarrow \boldsymbol \lambda_2^{(k)} + \mu_2(\boldsymbol v^{(k+1)}- \boldsymbol \nabla \boldsymbol z^{(k+1)})$

__end while__ <br>
__Output: $\boldsymbol x_* \leftarrow \boldsymbol x_k$__

While this looks complicated, all the subproblems have relatively straight forward solutions:  

__while__ $\boldsymbol x_k$ not converged (k=1,2, ...) __do__ <br>
>$\boldsymbol x^{(k+1)} \leftarrow S(\Phi \boldsymbol z^{(k)} -  \frac{ \boldsymbol \lambda_1^{(k)}}{\mu _1}, \frac{\alpha}{\mu_1})$<br>
$\boldsymbol v^{(k+1)}_i \leftarrow S_2((\nabla \boldsymbol z^{(k)})_i -  \frac{(\boldsymbol \lambda_2^{(k)})_i}{\mu_2}, \frac{\beta}{\mu_2})$<br>
$\boldsymbol z^{(k+1)} \leftarrow M^{-1}\boldsymbol b$<br>
$\boldsymbol \lambda_1^{(k+1)} \leftarrow \boldsymbol \lambda_1^{(k)} + \mu_1(\boldsymbol x^{(k+1)}- \Phi \boldsymbol z^{(k+1)})$ <br>
$\boldsymbol \lambda_2^{(k+1)} \leftarrow \boldsymbol \lambda_2^{(k)} + \mu_2(\boldsymbol v^{(k+1)}- \boldsymbol \nabla \boldsymbol z^{(k+1)})$

__end while__ <br>
__Output: $\boldsymbol x_* \leftarrow \boldsymbol x_k$__

```python
#wavelet transforms
#The helper functions have been adapted from __https://inst.eecs.berkeley.edu/~ee123/sp18/__ and aid in visualizing
#results from pywt

def wavMask(dims, scale):
    sx, sy = dims
    res = np.ones(dims)
    NM = np.round(np.log2(dims))
    for n in range(int(np.min(NM)-scale+2)//2):
        res[:int(np.round(2**(NM[0]-n))), :int(np.round(2**(NM[1]-n)))] = \
        res[:int(np.round(2**(NM[0]-n))), :int(np.round(2**(NM[1]-n)))]/2
    return res

def imshowWAV(Wim, scale=1):
    plt.imshow(np.abs(Wim)*wavMask(Wim.shape, scale), cmap = plt.get_cmap('gray'))

def coeffs2img(LL, coeffs):
    LH, HL, HH = coeffs
    return np.vstack((np.hstack((LL, LH)), np.hstack((HL, HH))))

def unstack_coeffs(Wim):
    L1, L2 = np.hsplit(Wim, 2)
    LL, HL = np.vsplit(L1, 2)
    LH, HH = np.vsplit(L2, 2)
    return LL, [LH, HL, HH]

def img2coeffs(Wim, levels=3):
    LL, c = unstack_coeffs(Wim)
    coeffs = [c]
    for i in range(levels-1):
        LL, c = unstack_coeffs(LL)
        coeffs.insert(0,c)
    coeffs.insert(0, LL)
    return coeffs

def dwt2(im):
    coeffs = pywt.wavedec2(im, wavelet='db4', mode='per', level=3)
    Wim, rest = coeffs[0], coeffs[1:]    
    for levels in rest:
        Wim = coeffs2img(Wim, levels)
    return Wim

def idwt2(Wim):
    coeffs = img2coeffs(Wim, levels=3)
    return pywt.waverec2(coeffs, wavelet='db4', mode='per')
```

We now describe the general problem setup. The ADMM algorithm will be used to reconstruct a 2D image of someones brain using subsampled fourier measurements defined in subsequent code cells as M1 and M2. Some details of the setup:

(1) Measurements are subsampled fourier coefficients.

(2) Sparsity exists in the wavelet coefficients of the image.

(3) Sparsity exists in the pixel gradients of the image.


```python
# The brain image below is the "ground truth" and referred to as 'z' in the main text.
# Our goal will be to recover this image from a relatively small number of measurements in the Fourier domain.
# Such recovery is possible due to L1 theory and the inherent sparsity in the image (when viewed in the appropriate basis).

data = np.load('brain.npz')
im, mask_unif, mask_vardens, pdf_unif, pdf_vardens = \
data['im'], data['mask_unif'], data['mask_vardens'], data['pdf_unif'], data['pdf_vardens'],

plt.figure(figsize=(8, 8));
plt.imshow(np.abs(im), cmap = 'gray')
plt.title('a brain')
plt.show()
```

```python
W_coeffs = dwt2(im)
plt.figure(figsize=(8, 8));
plt.imshow(np.abs(W_coeffs), cmap = 'gray') 
plt.show()
# The following plot demonstrates that the wavelet coefficients are quite sparse.
```

```python
def fft2c(x):
    return 1 / np.sqrt(np.prod(x.shape)) * np.fft.fftshift(np.fft.fft2(np.fft.ifftshift(x)))

def ifft2c(y):
    return np.sqrt(np.prod(y.shape)) * np.fft.ifftshift(np.fft.ifft2(np.fft.fftshift(y)))
```

```python
M = fft2c(im); # The raw measurements are the Fourier coefficents of the image (in k-space)

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(13, 3), ncols=3)

f1 = ax1.imshow(np.log(np.abs(M)), cmap = 'gray')
ax1.set_title('fully sampled fourier space')
fig.colorbar(f1, ax=ax1)

f2 = ax2.imshow(mask_unif, cmap = 'binary')
ax2.set_title('uniform mask')
fig.colorbar(f2, ax=ax2)

f3 = ax3.imshow(mask_vardens, cmap = 'binary')
ax3.set_title('variable density mask')
fig.colorbar(f3, ax=ax3)
plt.show()
```

In real-world applications, we would not have access to the fully sampled Fourier space. 
Rather, we would be able to probe the system at selected points in $k$-space, i.e., $(k_x, k_y)$.
We use these masks to generate two different types of measurements and we compare the recovery based on these different measurement schemes. Note that the masks specify the spatial frequencies $(k_x,k_y)$ at which we have recorded measurements. To simulate this procedure, these measurements are directly observed from the fully sampled Fourier space (which we do not have access to in practice). This is done below.

```python
M1 = (M * mask_unif) / pdf_unif; #subsampled measurements using the uniform mask
M2 = (M * mask_vardens)/ pdf_vardens; #subsampled measurements using the variable density mask
```

Note that the matrices $M_1$ and $M_2$ are still dense by construction as we are effectively just zeroing out
the Fourier coefficients at the unmeasured spatial frequencies.

Additionally, note that in the above figures, the variable density mask concentrates measurements 
in the lower (spatial) frequency range (near the origin). As evidenced by the leftmost figure,
this is also the region of highest Fourier coefficients for our brain image. 
With this mask, we retain more of the influential Fourier coefficients.

Moreover, as a benchmark to compare our sparse recovery algorithm, consider the recovered images that result result from simply applying the inverse Fourier transform to the above measurements. What is lost in this recovery scheme is the information stored in the unmeasured fourier coefficients, which we have simply just zeroed out. We have not yet leveraged the sparsity in the wavelet domain or in the pixel gradients.  

```python
M1_ifft = ifft2c(M1)
M2_ifft = ifft2c(M2)

fig, (ax1, ax2, ax3) = plt.subplots(figsize=(13, 3), ncols=3)

f1 = ax1.imshow(np.abs(im), cmap = 'gray')
ax1.set_title('original')
fig.colorbar(f1, ax=ax1)

f2 = ax2.imshow(np.abs(M1_ifft), cmap = 'gray')
ax2.set_title('M1, inverse FFT recovery')
fig.colorbar(f2, ax=ax2)

f3 = ax3.imshow(np.abs(M2_ifft), cmap = 'gray')
ax3.set_title('M2, inverse FFT recovery')
fig.colorbar(f3, ax=ax3)
plt.show()
```

Can we improve on this by leveraging sparsity? 

With the problem setup fully described, we now code our ADMM implementation combining pixel gradient sparsity and wavelet coefficient sparsity. Our specific implementation and notation is primarily based on the main text. Additional implementation details can be found in the following reference:

__Yang, J., Zhang, Y. and Yin, W. "A fast alternating direction method for TVL1-L2 signal reconstruction from partial Fourier data." IEEE Journal of Selected Topics in Signal Processing 4.2 (2010): 288-297.__

```python
## Recall the glossary from the main text:
# x = wavelet coefficients
# z = image in pixel space
# y = measure fourier coefficients
```

```python
# shrinkage/soft-thresholding operators 
# note that the operations are applied elementwise to vectors and matrices

def soft(w,t):
    return np.sign(w) * np.maximum(np.abs(w)-t,0) 

def soft2(w,t):
    wnorm = np.linalg.norm(w)
    if wnorm > 0:
        sol=w/(np.linalg.norm(w)) * np.maximum(np.linalg.norm(w)-t,0)
    else:
        sol=np.zeros(w.shape)
    return sol
```

```python
#derivative operators and helper functions

def DzDx(z):
    # z is an n-by-n image, x-derivatives computed using forward difference
    return np.hstack((-z[:,0:-1]+z[:,1::], np.atleast_2d(-z[:,-1] + z[:,0]).T))

def DzDy(z):
    # z is an n-by-n image, y-derivatives computed using forward difference
    return np.vstack((-z[0:-1,:]+z[1::,:], -z[-1,:] + z[0,:]))

def Dtv(v):
    nrows = len(v)
    ncols = len(v[0])
    Dtv = np.zeros((nrows,ncols))
    
    for i1 in range(nrows):
        for i2 in range(ncols):
            Dtv[i1,i2] = (v[i1][i2][0] - v[i1][np.mod(i2+1, ncols)][0]) + (v[i1][i2][1] - v[np.mod(i1+1, nrows)][i2][1])
            
    return Dtv
```

```python
from tqdm.auto import tqdm
# this implementation is currently not complete! It also needs to be properly tuned...

def admm(y, opts):
    nrows = y.shape[0] #dimensions of the Fourier space
    ncols = y.shape[1]
    m = np.count_nonzero(y) #number of actual measurements in y
    N = nrows*ncols
    #tuning parameters
    alpha=100#weights wavelet sparsity
    beta=20 #weights gradient sparsity
    mu1=0.1
    mu2=0.1
    mask = (np.abs(y) > 0).astype(int)
    ##initializations
    
    #image
    z = np.zeros((nrows,ncols))
    #wavelet coefficients
    x = np.zeros((nrows,ncols))
    #pixel gradients
    v = [[np.zeros(2,) for i1 in range(nrows)] for i2 in range(ncols)]
    
    #multipliers
    lam1 = np.zeros((nrows,ncols))
    lam2 = v.copy()
    
    # Compute the Mhat vector (see references)
    
    #kernels of the finite difference operators
    D_dx = np.zeros((nrows,ncols))
    D_dx[0,0] = -1
    D_dx[0,1] = 1

    D_dy = np.zeros((nrows,ncols))
    D_dy[0,0] = -1
    D_dy[1,0] = 1
    
    Dhat_T_Dhat_dx = np.square(np.abs(fft2c(-D_dx)))
    Dhat_T_Dhat_dy = np.square(np.abs(fft2c(-D_dy)))

    Mhat = mu2*(Dhat_T_Dhat_dx+Dhat_T_Dhat_dy)+mask+mu1
    
    ## optimization
    for ii in tqdm(range(opts['maxIter'])):
        
        x = soft(dwt2(z)-lam1/mu1,alpha/mu1)
        
        D_x = DzDx(np.abs(z))
        D_y = DzDy(np.abs(z))
        
        v = [[soft2( np.asarray([ D_x[i1,i2], D_y[i1,i2]]) - lam2[i1][i2]/mu2, beta/mu2 )  for i1 in range(nrows)] for i2 in range(ncols)]
        yhat = fft2c( Dtv(v) + idwt2(x) ) + y
                
        z = ifft2c(yhat / Mhat )
        
        D_x = DzDx(np.abs(z))
        D_y = DzDy(np.abs(z))
        
        ## update multipliers
        lam1 = lam1 + mu1*(x-dwt2(z))
        lam2 = [[ lam2[i1][i2] + mu2*(v[i1][i2]-np.asarray([D_x[i1,i2], D_y[i1,i2]]))  for i1 in range(nrows)] for i2 in range(ncols)]
        
    return z
```

We first investigate the recovery properties assuming the measurements come from the uniform mask of the fourier coefficients, i.e. $M_1$ is the measured quantity. Hence, we do not observe all of the fourier coefficients, but only some of them.

```python
opts = {'tol': 1e-4, 'maxIter': 100} #currently have not implemented tolerance convergence
z1 = admm(M1, opts)
z2 = admm(M2, opts)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize=(13, 3), ncols=3)

f1 = ax1.imshow(np.abs(im), cmap = 'gray')
ax1.set_title('original')
fig.colorbar(f1, ax=ax1)

f2 = ax2.imshow(np.abs(z1), cmap = 'gray')
ax2.set_title('M1, ADMM recovery')
fig.colorbar(f2, ax=ax2)

f3 = ax3.imshow(np.abs(z2), cmap = 'gray')
ax3.set_title('M2, ADMM recovery')
fig.colorbar(f3, ax=ax3)
plt.show()
```
