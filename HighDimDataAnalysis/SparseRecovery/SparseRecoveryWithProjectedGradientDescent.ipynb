

## Projected subgradient method for sparse recovery with $\ell_1$ minimization

The projected gradient descent algorithm repeats the following iteration until convergence:

First, a (sub)gradient update,

$$\boldsymbol z_{k} = \boldsymbol x_k - t_k \nabla f(\boldsymbol x_k) $$

followed by a projection step onto the constraint set $C$,

$$\boldsymbol x_{k+1} = P_C[\boldsymbol z_k  ] $$ 

Recall, the projection is defined as,

$$ P_C[\boldsymbol z] = \text{arg} \min_{\boldsymbol x\in C} \frac{1}{2} \| \boldsymbol z- \boldsymbol x \|_2^2 = h(\boldsymbol x)$$.


```python
def projected_gradient_descent(A,y, xtol, maxIter):
    m = A.shape[0]
    n = A.shape[1]
    At = np.transpose(A)
    AAt = np.matmul(A,At)
    AtinvAAt = np.matmul(At,np.linalg.inv(AAt))
    P = np.identity(n) - np.matmul(AtinvAAt,A)
    xtilde = AtinvAAt.dot(y) 
    
    xOld=np.zeros(n)
    t=0
    gtol = xtol+1
    
    while t <= maxIter or  gtol > xtol:
        t = t+1
        xNew = xtilde + P.dot(xOld - 1/t* np.sign(xOld))
        xOld = xNew
        gtol = np.linalg.norm(xNew-xOld)
    
    return xNew
```

We will now use the projected subgradient algorithm to solve

$$ \text{minimize} \|\boldsymbol x\|_1$$
$$ \text{subject to  } \boldsymbol y= \boldsymbol A \boldsymbol x   $$

for randomly generated instance of the problem. 

```python
# Problem setup

n = 100           # Matrix size
m = 50
k = 10            # Number of nonzeros, specifies the sparsity of the solution 

x = np.concatenate((np.random.randn(k),np.zeros(n-k))) # Create sparse vector 
x = x[np.random.permutation(n)];

A = np.random.randn(m,n)                   # Create random measurement matrix A 

y = A.dot(x)     # measurement vector, this is all we observe of x

plt.figure(figsize=(15, 4));
plt.subplot(1,3,1), plt.imshow(A); plt.title('A matrix');
plt.subplot(1,3,2), plt.stem(y); plt.title('y');
plt.subplot(1,3,3), plt.stem(x); plt.title('x');
```

```python
# Problem solution

xest = projected_gradient_descent(A,y,1e-9,1e4)
error = np.absolute(xest-x) #absolute value of error

plt.subplot(1,2,1), plt.stem(error); plt.title('error');
plt.subplot(1,2,2), plt.stem(xest); plt.title('recovered');
```

Below we fix the number of iterations and run the algorithm on a randomly generated problem for increasing levels of sparsity in $x$. At what level of sparsity does the recovery begin to fail?

```python
n = 100           # Matrix size
m = 50

A = np.random.randn(m,n)                   # Create Matrix A 
k = range(1,n+1)
error = np.zeros(len(k))

for ii in range(len(k)):
    x = np.concatenate((np.random.randn(k[ii]),np.zeros(n-k[ii]))) # Create sparse vector 
    x = x[np.random.permutation(n)];
    y = A.dot(x)     
    numIter = 200
    tol = 0
    xest = projected_gradient_descent(A,y,tol,numIter)
    error[ii] = np.linalg.norm(xest-x)
    
```

```python
plt.plot(k,error) 
plt.title('sparsity vs. recovery error')
plt.xlabel('number non-zeros, k')
plt.ylabel('reconstruction error (1-norm)');

# as observed in the figure below, the recovery error starts rising at around k=17.
```

## Phase transition for sparse recovery via projected gradient descent

Here we sweep over a variety of problem instances to investigate the general behavior of the algorithms performance.

Note that the stopping criteria that I've implemented is based on the euclidean distance between sequential iterates of the projected gradient descent algorithm. 

First, we vary the sparsity $k$ of $\boldsymbol x$, and plot the fraction of successful recoveries over 100 trials / randomly generated problem instances. 

```python
#this takes a bit of time to run

n = 100           # Matrix size
m = 50
k = range(1,n+1)         # Number of nonzeros 
numTrials = 100    

sfrac = np.zeros(len(k))

for ii in range(len(k)):
    count = 0 #keeps track of successful trials
    for jj in range(numTrials):
        A = np.random.randn(m,n)
        x = np.concatenate((np.random.randn(k[ii]),np.zeros(n-k[ii]))) # Create sparse vector 
        x = x[np.random.permutation(n)];
        y = A.dot(x)     
        numIter = 10000
        tol = 1e-5
        xest = projected_gradient_descent(A,y,tol,numIter)
        trialError = np.linalg.norm(xest-x)
        if trialError < 1e-3:
            count=count+1
    sfrac[ii] = count/numTrials
    
plt.plot(k,sfrac); plt.xlabel('k (sparsity)'); plt.ylabel('fraction success')
```

Next, we fix the dimensionality of the ambient space to $n=50$ and sweep through increasing numbers of measurements $m$ and sparsity $k$ such that:

(a) the ratio $\frac{m}{n}$ spans [0,1] 

(b) $\eta = \frac{k}{n}$ between [0,1]

The results are best visualized in a phase transition plot. 

```python
n = 50
delta = np.linspace(0,1,20)
eta = np.linspace(0,1,20)
Delta, Eta = np.meshgrid(delta,eta)
Sfrac = np.zeros(Delta.shape)
numTrials = 100

for ii in range(len(delta)):
    for jj in range(len(eta)):
        m = Delta[ii,jj]*n
        k = Eta[ii,jj]*n
        m = m.astype(np.int64)
        k = k.astype(np.int64)
        count = 0 #keeps track of successful trials
        for tt in range(numTrials):
            A = np.random.randn(m,n)
            x = np.concatenate((np.random.randn(k),np.zeros(n-k))) # Create sparse vector 
            x = x[np.random.permutation(n)];
            y = A.dot(x)     
            numIter = 10000
            tol = 1e-5
            xest = projected_gradient_descent(A,y,tol,numIter)
            trialError = np.linalg.norm(xest-x)
            if trialError < 1e-3:
                count=count+1
        Sfrac[ii,jj] = count/numTrials
```

```python
fig = plt.figure(figsize=(10, 5));
cs = plt.contourf(Delta,Eta, Sfrac)
fig.colorbar(cs)
plt.xlabel(r'$\delta = m/n$'),  plt.ylabel(r'$\eta = k/n$'), 
plt.title('phase transition curve'); 
```

The $\ell_1$ minimization succeeds where the phase transition region is near 1, i.e., the lower right region of the plot. In this region, all of the trials demonstrate successful recovery. In the upper left region, we observe that none of the trials are successful and the $\ell_1$ strategy fails. The in-between region is where the trial-to-trial variability (due to the randomly sampled setup to create the ground truth) leads to both successes and failures. From the phase transition curve we see that as $m \rightarrow n$ (we are allowed more "measurements") we can recover vectors that are less sparse ($\eta >> 0$). The plot indicates that we can always expect to recover very sparse vectors ($\eta << 1$).  

## From $\ell_{1}$ to  $\ell_{\infty}$ minimization

Here we use the subdifferential of the $\|\boldsymbol x\|_\infty$ norm to solve the optimization problem:

$$\min \| \boldsymbol x\|_\infty \quad \mbox{subject to} \quad \boldsymbol y = A \boldsymbol x.$$


```python
def ebv(n,i): #generate the euclidean basis vector ei
    e = np.zeros(n)
    e[i] = 1
    return e

def maxSupp(x): #returns all indices containing the maximum magnitude element of x 
    maxElem = np.amax(np.absolute(x))
    idx = []
    for kk in range(len(x)):
        if np.absolute(x[kk]) == maxElem:
            idx.append(kk)
    return idx

def subgrad(x): #subgradient of the infinity norm. See problem 2.1 of the written assignment for the calculation.
    n = len(x)
    argmaxIdx = maxSupp(x) 
    a = np.random.rand(n)
    a = a / np.sum(a)
    g = 0;
    for kk in range(len(argmaxIdx)):
        i = argmaxIdx[kk]
        ei = ebv(n,i)
        g = g + a[kk]*ei*np.sign(x[i]) # pick from convex hull
    return g

def pgd_Linf(A,y, xtol, maxIter):
    m = A.shape[0]
    n = A.shape[1]
    At = np.transpose(A)
    AAt = np.matmul(A,At)
    AtinvAAt = np.matmul(At,np.linalg.pinv(AAt))
    P = np.identity(n) - np.matmul(AtinvAAt,A) #projection operator is unchanged from before
    xtilde = AtinvAAt.dot(y) #projection operator is unchanged
    
    xOld=np.zeros(n)
    t=0
    tol = xtol+1
    
    while t <= maxIter or  tol > xtol:
        t = t+1
        g = subgrad(xOld)
        #pdb.set_trace() #debugging
        xNew = xtilde + P.dot(xOld - 1/t*g) #move in the negative gradient direction and project.
        xOld = xNew
        tol = np.linalg.norm(xNew-xOld)
    
    return xNew
```

```python
# In order to show that the algorithm is working, we will investigate a simple example where we know the solution. 
# Does the above algorithm recover the correct solution?
A = np.array([[3,3], [2,2]]) #null(A) = [1,-1]
x0 = np.ones(2) #true answer
y = A.dot(x0)

xest = pgd_Linf(A,y, 1e-9, 10000)
print(xest )
#Answer is close to the correct answer of (1,1), but convergence appears slow....
#So, we see that in this simple example, the implemented projected gradient descent algorithm for 
#$\ell_\infty$ optimization work. 
```
